## 영상 크기 조정

```python
r,c = img.shape[:2]
new_img = cv.resize(img, (2*r, 2*c))
plt.imshow(new_img)
plt.plot()
```

## 영상 영역의 이동

```python
M = np.float32([[1, 0, 100], [0, 1, 100]])
new_img = cv.warpAffine(img, M, (c, r))
plt.imshow(new_img)
plt.plot()
```

```python
def color_rect():

    # Color rectangle 생성
    g = np.zeros((256, 256, 3), np.uint8)
    for i in range(256):
        g[i,:,:] = (0, i, 255 - i)

    return g

g = color_rect()

col = cv.cvtColor(g, cv.COLOR_BGR2RGB)
plt.imshow(col)
plt.show()
```
